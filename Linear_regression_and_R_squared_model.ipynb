```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from statistics import mean
import random
style.use('fivethirtyeight')
```

```
#xs=np.array([1,2,3,4,5,6])
#ys=np.array([5,4,6,5,6,7])
def best_fit_slope_and_intercept(xs,ys):
  m=((mean(xs)*mean(ys) -mean(xs*ys) ) / ((mean(xs)**2)-mean(xs**2) ))
  b=mean(ys)-(m*mean(xs))
  return m,b
def create_dataset(num, var, step , corelation='False'):
  ys=[]
  val=1
  for i in range(num):
    y= val+ random.randrange(-var, var)
    ys.append(y)
    if corelation and corelation == 'pos':
      val+=step
    elif corelation and corelation == 'neg':
      val-=step

  xs= [i for i in range(len(ys))]
  return (np.array(xs,dtype= np.float64),np.array(ys,dtype= np.float64))

def squared_error(y_orig, y_act):
  return sum((y_orig-y_act)**2)

```

```
xs , ys=create_dataset(150,50,2,'neg')

m,b = best_fit_slope_and_intercept(xs,ys)
regression_line = [(m*x +b) for x in xs]

def R_squared_value(ys,regression_line):
  k=mean(ys)
  r=1-(squared_error(ys,regression_line)/squared_error(k,regression_line))
  return r
distribution_cofficient = R_squared_value(ys,regression_line)
print(distribution_cofficient)

plt.scatter(xs,ys)
plt.plot(xs,regression_line)
plt.show()
```
